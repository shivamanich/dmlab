```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/diabetes_young_adults_india.csv'
data = pd.read_csv(file_path)
print(data)
```

```python

# Step 1: Correlation Analysis
# Select only numerical columns for correlation
numerical_data = data.select_dtypes(include=['int64', 'float64'])
# Compute correlation matrix
correlation_matrix = numerical_data.corr()
# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Step 2: Visualize Distributions
# Plot distributions for numerical variables
numerical_columns = ['Age', 'BMI', 'Fasting_Blood_Sugar', 'HbA1c',
                     'Cholesterol_Level', 'Sleep_Hours', 'Stress_Level', 'Screen_Time']
data[numerical_columns].hist(figsize=(14, 10), bins=30, color='skyblue', edgecolor='black')
plt.suptitle("Distributions of Numerical Variables")
plt.show()
```

```python
# Step 3: Scatter Plot for Trends
# Example: BMI vs Fasting Blood Sugar
plt.figure(figsize=(8, 6))
sns.scatterplot(x='BMI', y='Fasting_Blood_Sugar', hue='Diabetes_Type', data=data, alpha=0.6)
plt.title("BMI vs Fasting Blood Sugar by Diabetes Type")
plt.xlabel("BMI")
plt.ylabel("Fasting Blood Sugar")
plt.show()
```

```python
# Step 4: Box Plot for Outliers
# Example: BMI by Diabetes Type
plt.figure(figsize=(8, 6))
sns.boxplot(x='Diabetes_Type', y='BMI', data=data)
plt.title("BMI Distribution by Diabetes Type")
plt.xlabel("Diabetes Type")
plt.ylabel("BMI")
plt.show()
```

```python
# Step 5: Categorical Analysis
# Count plot for Diabetes_Type by Gender
plt.figure(figsize=(8, 6))
sns.countplot(x='Diabetes_Type', hue='Gender', data=data)
plt.title("Diabetes Type by Gender")
plt.xlabel("Diabetes Type")
plt.ylabel("Count")
plt.show()
```
