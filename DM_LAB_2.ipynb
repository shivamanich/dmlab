```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (replace 'dataset.csv' with your file path if running locally)
file_path = 'dataset.csv'
df = pd.read_csv('dataset.csv', delimiter=';')

# Display statistical summaries
print("Statistical Summaries:")
print(df.describe(include='all'))

# 1. Line Plot: Fixed Acidity Trend
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['fixed acidity'], marker='o', linestyle='-', color='blue', alpha=0.7, label='Fixed Acidity')
plt.title('Line Plot of Fixed Acidity')
plt.xlabel('Index')
plt.ylabel('Fixed Acidity')
plt.legend()
plt.grid()
plt.show()

# 2. Scatter Plot: Alcohol vs. Quality
plt.figure(figsize=(8, 6))
sns.scatterplot(x='alcohol', y='quality', data=df, alpha=0.7)
plt.title('Scatter Plot of Alcohol vs. Quality')
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.show()

# 3. Bar Plot: Average Alcohol by Quality
plt.figure(figsize=(8, 6))
sns.barplot(x='quality', y='alcohol', data=df, ci=None, palette='viridis')
plt.title('Bar Plot of Average Alcohol by Quality')
plt.xlabel('Quality')
plt.ylabel('Average Alcohol')
plt.show()

# 4. Histogram: Distribution of pH
plt.figure(figsize=(8, 6))
sns.histplot(df['pH'], kde=True, bins=20, color='purple')
plt.title('Histogram of pH')
plt.xlabel('pH')
plt.ylabel('Frequency')
plt.show()

# 5. Pie Chart: Distribution of Quality
plt.figure(figsize=(8, 8))
quality_counts = df['quality'].value_counts()
plt.pie(quality_counts, labels=quality_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Pie Chart of Quality Distribution')
plt.show()
```
